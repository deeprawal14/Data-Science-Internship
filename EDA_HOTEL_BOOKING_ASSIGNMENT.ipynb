# EDA (Exploratory Data Analysis)

## 1.Which type of hotel is mostly prefered by the guests?

```python
df['hotel'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['hotel'])
plt.xlabel('hotel')
plt.ylabel('Count')
plt.title("Preferred Hotel Type")
```

```python
plt.pie(df['hotel'].value_counts(),labels=df['hotel'].value_counts().index.tolist(),explode=[0.03, 0.03],
        autopct='%1.1f%%',colors=[ 'orange','blue'],
        radius=1,textprops={'fontsize': 12, 'fontweight': 'medium'})

plt.show()
```

### Observation City Hotel is most preffered hotel by guests. Thus city hotels has maximum bookings which is 66.4%

## 2.Which type of meal is mostly preferred by the guests?

```python
df.pivot_table(index = 'hotel', columns = 'meal', values = 'total_guests', aggfunc = 'sum')
```

```python
plt.figure( figsize=(10, 8))
sns.countplot(x = df['meal'])
plt.title("Most Preffered Meal ",loc='center')
plt.show()
```

### Observation- Most preferred meal type is BB (Bed and breakfast)

## 3.Which Agent made the most bookings?

```python
most_bookings= df.groupby(['agent'])['agent'].agg({'count'}).reset_index().rename(columns={'count': "Most Bookings count" }).sort_values(by='Most Bookings count',ascending=False)
```

```python
most_bookings.drop(most_bookings[most_bookings['agent']==0].index,inplace=True)
```

```python
top_bookings=most_bookings[:5]
```

```python
plt.figure(figsize = (8,5))
sns.barplot(x =top_bookings['agent'], y =top_bookings['Most Bookings count'],palette='Spectral',order=top_bookings['agent'])
plt.title('Most Bookings Count vs Agent ',loc='center',pad=30,fontdict={'fontsize': 20, 'fontweight': 'medium'})
plt.show()
```

### Observations:- Agent number 9 made the most booking more than 25k.

## 4.In which month most of the bookings happened?

```python
plt.figure( figsize=(15, 9))
sns.countplot(x= df['arrival_date_month'])
plt.title("Booking month vs number of bookings ")
plt.ylabel("number of bookings")
plt.show()
```

###  Observations:- August has the most bookings

## 5.Which Hotels has the most repeated guests?

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['is_repeated_guest'],hue=df['hotel'])
plt.title(" repeated guests vs Hotel ")
plt.xlabel(' Repeated Guests')
plt.ylabel('counts')
plt.legend(loc=(.01,.9))
plt.show()
```

```python
df1=df[df['is_repeated_guest']==1]
plt.figure(figsize=(10,5))
data = df1.groupby("hotel")["is_repeated_guest"].count()
colors = ['blue','orange']
explode=(0,0.2)
data.plot.pie(autopct="%.1f%%",colors=colors,explode=explode);

plt.tight_layout()
```

### Observations:- It can be seen that there is not much difference in numbers and they tend to have same number of repeated guests.

## 6.From which country the most guests are coming?

```python
country_df = df['country'].value_counts().reset_index().rename(columns={'index': 'country','country': 'count of guests'})[:10]

plt.figure(figsize=(20, 8))
sns.barplot(x=country_df['country'], y=country_df['count of guests'],palette='tab20')
plt.title("Number of guests from different Country",loc='center',pad=30,fontdict={'fontsize': 20, 'fontweight': 'medium'})
plt.show()
```

### Observations:- Most guests are from Portugal.

## 10.What is the pecentage of cancellation for each hotel?

```python
total_registered_hotels = df.pivot_table(index = 'hotel', values = 'is_canceled', aggfunc = 'count')
total_registered_hotels.reset_index().rename(columns = {'is_canceled': 'Registered_Hotels'})
```

```python
total_canceled_hotels = df[df.is_canceled == 1].pivot_table(index = 'hotel', values = 'is_canceled', aggfunc = 'count')
total_canceled_hotels.reset_index().rename(columns = {'is_canceled': 'Canceled_Hotels'})
```

```python
percent_canceled_hotels = (total_canceled_hotels/total_registered_hotels) * 100
percent_canceled_hotels.reset_index().rename(columns = {'hotel': 'Hotel_Type', 'is_canceled': 'Percent_of_Canceled_Bookings '})
```

```python
df1=df[df['is_canceled']==1]
plt.figure(figsize=(10,5))
data = df1.groupby("hotel")["is_canceled"].count()
colors = ['blue','orange']
explode=(0,0.2)
data.plot.pie(autopct="%.1f%%",colors=colors,explode=explode);

plt.tight_layout()
```

### Observations:- Cancellation rate in City hotels =74.8%. Cancellation rate in Resort hotels= 25.2%. Cancellation is less in Resort hotels as compared to City hotels.
